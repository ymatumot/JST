# 課題３：惑星軌道計算

```python
import numpy as np
import matplotlib.pyplot as plt

x  = 2.0
#x = 1.0
y  = 0.0
vx = 0.0
vy = 0.7
#vy = 1.4

#ma  = 0.8
ma = 0.5
mb  = 1.0-ma
xa  = -mb
ya  =  0.0
vxa =  0.0
vya = -mb
xb  =  ma
yb  =  0.0
vxb =  0.0
vyb =  ma

dt  = 0.01
tmax = 50.0

k0 = np.zeros(4)
k1 = np.zeros(4)
k2 = np.zeros(4)
k3 = np.zeros(4)

t=0
while t < tmax:
  rb = np.sqrt((xa-xb)*(xa-xb)+(ya-yb)*(ya-yb))
  rb3= rb*rb*rb
  k0[0]= dt*vxa
  k0[1]= dt*vya
  k0[2]= dt*(-mb*(xa-xb)/rb3)
  k0[3]= dt*(-mb*(ya-yb)/rb3)
  x1 = xa+k0[0]/2.0
  y1 = ya+k0[1]/2.0
  vx1=vxa+k0[2]/2.0
  vy1=vya+k0[3]/2.0
  rb = np.sqrt((x1-xb)*(x1-xb)+(y1-yb)*(y1-yb))
  rb3= rb*rb*rb
  k1[0]= dt*vx1
  k1[1]= dt*vy1
  k1[2]= dt*(-mb*(x1-xb)/rb3)
  k1[3]= dt*(-mb*(y1-yb)/rb3)
  x2 = xa+k1[0]/2.0
  y2 = ya+k1[1]/2.0
  vx2=vxa+k1[2]/2.0
  vy2=vya+k1[3]/2.0
  rb = np.sqrt((x2-xb)*(x2-xb)+(y2-yb)*(y2-yb))
  rb3= rb*rb*rb
  k2[0]= dt*vx2
  k2[1]= dt*vy2
  k2[2]= dt*(-mb*(x2-xb)/rb3)
  k2[3]= dt*(-mb*(y2-yb)/rb3)
  x3 = xa+k2[0]
  y3 = ya+k2[1]
  vx3=vxa+k2[2]
  vy3=vya+k2[3]
  rb = np.sqrt((x3-xb)*(x3-xb)+(y3-yb)*(y3-yb))
  rb3= rb*rb*rb
  k3[0]= dt*vx3
  k3[1]= dt*vy3
  k3[2]= dt*(-mb*(x3-xb)/rb3)
  k3[3]= dt*(-mb*(y3-yb)/rb3)
  xan =xa +(k0[0]+2.0*k1[0]+2.0*k2[0]+k3[0])/6.0
  yan =ya +(k0[1]+2.0*k1[1]+2.0*k2[1]+k3[1])/6.0
  vxan=vxa+(k0[2]+2.0*k1[2]+2.0*k2[2]+k3[2])/6.0
  vyan=vya+(k0[3]+2.0*k1[3]+2.0*k2[3]+k3[3])/6.0

  ra = np.sqrt((xb-xa)*(xb-xa)+(yb-ya)*(yb-ya))
  ra3= ra*ra*ra
  k0[0]=dt*vxb
  k0[1]=dt*vyb
  k0[2]=dt*(-ma*(xb-xa)/ra3)
  k0[3]=dt*(-ma*(yb-ya)/ra3)
  x1 =xb+k0[0]/2.0
  y1 =yb+k0[1]/2.0
  vx1=vxb+k0[2]/2.0
  vy1=vyb+k0[3]/2.0
  ra =np.sqrt((x1-xa)*(x1-xa)+(y1-ya)*(y1-ya))
  ra3= ra*ra*ra
  k1[0]=dt*vx1
  k1[1]=dt*vy1
  k1[2]=dt*(-ma*(x1-xa)/ra3)
  k1[3]=dt*(-ma*(y1-ya)/ra3)
  x2 =xb+k1[0]/2.0
  y2 =yb+k1[1]/2.0
  vx2=vxb+k1[2]/2.0
  vy2=vyb+k1[3]/2.0
  ra =np.sqrt((x2-xa)*(x2-xa)+(y2-ya)*(y2-ya))
  ra3=ra*ra*ra
  k2[0]=dt*vx2
  k2[1]=dt*vy2
  k2[2]=dt*(-ma*(x2-xa)/ra3)
  k2[3]=dt*(-ma*(y2-ya)/ra3)
  x3 =xb+k2[0]
  y3 =yb+k2[1]
  vx3=vxb+k2[2]
  vy3=vyb+k2[3]
  ra =np.sqrt((x3-xa)*(x3-xa)+(y3-ya)*(y3-ya))
  ra3=ra*ra*ra
  k3[0]=dt*vx3
  k3[1]=dt*vy3
  k3[2]=dt*(-ma*(x3-xa)/ra3)
  k3[3]=dt*(-ma*(y3-ya)/ra3)
  xbn =xb +(k0[0]+2.0*k1[0]+2.0*k2[0]+k3[0])/6.0
  ybn =yb +(k0[1]+2.0*k1[1]+2.0*k2[1]+k3[1])/6.0
  vxbn=vxb+(k0[2]+2.0*k1[2]+2.0*k2[2]+k3[2])/6.0
  vybn=vyb+(k0[3]+2.0*k1[3]+2.0*k2[3]+k3[3])/6.0

  ra = np.sqrt((x-xa)*(x-xa)+(y-ya)*(y-ya))
  rb = np.sqrt((x-xb)*(x-xb)+(y-yb)*(y-yb))   
  ra3= ra*ra*ra
  rb3= rb*rb*rb
  k0[0]=dt*vx
  k0[1]=dt*vy
  k0[2]=dt*(-ma*(x-xa)/ra3-mb*(x-xb)/rb3)
  k0[3]=dt*(-ma*(y-ya)/ra3-mb*(y-yb)/rb3)
  x1 = x+k0[0]/2.0
  y1 = y+k0[1]/2.0
  vx1= vx+k0[2]/2.0
  vy1= vy+k0[3]/2.0
  ra = np.sqrt((x1-xa)*(x1-xa)+(y1-ya)*(y1-ya))
  rb = np.sqrt((x1-xb)*(x1-xb)+(y1-yb)*(y1-yb))
  ra3= ra*ra*ra
  rb3= rb*rb*rb
  k1[0]=dt*vx1
  k1[1]=dt*vy1
  k1[2]=dt*(-ma*(x1-xa)/ra3-mb*(x1-xb)/rb3)
  k1[3]=dt*(-ma*(y1-ya)/ra3-mb*(y1-yb)/rb3)
  x2 = x+k1[0]/2.0
  y2 = y+k1[1]/2.0
  vx2= vx+k1[2]/2.0
  vy2= vy+k1[3]/2.0
  ra = np.sqrt((x2-xa)*(x2-xa)+(y2-ya)*(y2-ya))
  rb = np.sqrt((x2-xb)*(x2-xb)+(y2-yb)*(y2-yb))
  ra3= ra*ra*ra
  rb3= rb*rb*rb
  k2[0]=dt*vx2
  k2[1]=dt*vy2
  k2[2]=dt*(-ma*(x2-xa)/ra3-mb*(x2-xb)/rb3)
  k2[3]=dt*(-ma*(y2-ya)/ra3-mb*(y2-yb)/rb3)
  x3 =x+k2[0]
  y3 =y+k2[1]
  vx3=vx+k2[2]
  vy3=vy+k2[3]
  ra = np.sqrt((x3-xa)*(x3-xa)+(y3-ya)*(y3-ya))
  rb = np.sqrt((x3-xb)*(x3-xb)+(y3-yb)*(y3-yb))
  ra3= ra*ra*ra
  rb3= rb*rb*rb
  k3[0]=dt*vx3
  k3[1]=dt*vy3
  k3[2]=dt*(-ma*(x3-xa)/ra3-mb*(x3-xb)/rb3)
  k3[3]=dt*(-ma*(y3-ya)/ra3-mb*(y3-yb)/rb3)
  x  = x +(k0[0]+2.0*k1[0]+2.0*k2[0]+k3[0])/6.0
  y  = y +(k0[1]+2.0*k1[1]+2.0*k2[1]+k3[1])/6.0
  vx = vx+(k0[2]+2.0*k1[2]+2.0*k2[2]+k3[2])/6.0
  vy = vy+(k0[3]+2.0*k1[3]+2.0*k2[3]+k3[3])/6.0
  xa=xan
  ya=yan
  xb=xbn
  yb=ybn
  vxa=vxan
  vya=vyan
  vxb=vxbn
  vyb=vybn
  t=t+dt
  plt.plot(xa,ya,'+',markersize=1,c='blue')
  plt.plot(xb,yb,'x',markersize=1,c='red')
  plt.plot(x,y,'o',markersize=1,c='green')
plt.xlim(-3.0,3.0)
plt.ylim(-3.0,3.0)
plt.show()
```
